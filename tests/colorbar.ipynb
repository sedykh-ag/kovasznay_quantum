```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [np.random.randn(100, 3) for _ in range(2)]
```

```python
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(16, 9))

for i in range(2):
    mappable = axs[i].scatter(x[i][:, 0], x[i][:, 1], c=x[i][:, 2])
    fig.colorbar(mappable, ax=axs[i], location="bottom", shrink=0.83)

    axs[i].set_aspect("equal")

fig.show()
```
